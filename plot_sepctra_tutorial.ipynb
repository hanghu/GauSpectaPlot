```python
import numpy as np
import pandas as pd
from spectra_tools import spectrum_boardening,plot_spectrum
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_fwf('dummy_spectra.dat',header=None)
```

```python
df
```

```python
E = df[0]
I = df[1]
```

## Supported Broadening Lineshape:

pass variable through `lineshape`
1. Gaussian
2. Lorentzian
3. Voigt (a convolution of Gaussian and Lorentzian)
4. self-defined function

```python
# FWHM as [FWHM_G, FWHM_L]
FWHM_G = 0.4
FWHM_L = 0.2 
Ex,Iy = spectrum_boardening(E, I, E_range=(705,715), resol=None, lineshape='V',
                            FWHM=[FWHM_G,FWHM_L], shift=0.0, normalize=False)
```

```python
fig, ax=plt.subplots(figsize=(12,8))
ax.plot(Ex,Iy)
```

## Or directly plot from `plot_spectrum`

```python
FWHM_G = 0.4
FWHM_L = 0.2 
fig, ax=plt.subplots(figsize=(12,8))
plot_spectrum(E, I, E_range=(705,715), output_resol=None, lineshape='V',
              FWHM=[FWHM_G,FWHM_L], ax=ax, shift=0.0, normalize=False)
```
